```python
import pandas as pd
import numpy as np
```

```python
with open('trans1.json','r') as f1:
    data1= (line.strip() for line in f1)
    data_json1="[{0}]".format(','.join(data1))
```

```python
data1 = pd.read_json(data_json1)
```

```python
data1.shape
```

```python
with open('trans2.json','r') as f2:
    data2= (line.strip() for line in f2)
    data_json2="[{0}]".format(','.join(data2))
```

```python
data2=pd.read_json(data_json2)
```

```python
data2.shape
```

```python
df= pd.merge(data1,data2,how='outer')
```

```python
df.head()
```

```python
#info shows no null entries, but there are clearly are some null/empty entries

df.info()
```

```python
#check for empty columns

empty_columns=df.loc[:, (df == '').all()]
empty_columns.head()
```

```python
#accountNumber and customerID are exactly the same

same_id=np.where(df["accountNumber"] == df["customerId"], True, False)
df[same_id==False]
```

```python
#dropping empty and duplicated columns


df=df.drop(columns=['accountNumber','echoBuffer','merchantCity','merchantState','merchantZip','posOnPremises','recurringAuthInd'])
```

```python
df.shape
```

```python
df.head(10)
```

```python
#9695 rows that contains at least one empty entry

df_empty=df[df.applymap(lambda x: len(str(x)) ==0).any(axis=1)]
df.head()
```

```python
#Most rows that contain empty entries contain only one with the max empty entries of 3
#Do not need to drop any rows for now

df_empty.iloc[:, 1:].eq("").sum(axis=1).describe()
```

```python
#acqCountry and posEntryMode contain a few empty entries, but relatively small for the data size
#Do not need to drop any of these columns


df_empty.iloc[:, 1:].eq("").sum(axis=0)
```

```python
df.info()
```

```python
#fix data types

df[['customerId','cardCVV','enteredCVV','cardLast4Digits']]=df[['customerId','cardCVV','enteredCVV','cardLast4Digits']].astype('object')
```

```python
#remove 'T' in order to split up by date and time

df['transactionDateTime']=df['transactionDateTime'].map(lambda x: str(x).replace('T',' '))
```

```python
df['date'] = pd.to_datetime(df['transactionDateTime']).dt.date
df['time'] = pd.to_datetime(df['transactionDateTime']).dt.time
df.head()
```

```python
col_names=['customerId', 'creditLimit', 'availableMoney', 'transactionDateTime', 'date', 'time',
       'transactionAmount', 'merchantName', 'acqCountry',
       'merchantCountryCode', 'posEntryMode', 'posConditionCode',
       'merchantCategoryCode', 'currentExpDate', 'accountOpenDate',
       'dateOfLastAddressChange', 'cardCVV', 'enteredCVV', 'cardLast4Digits',
       'transactionType', 'currentBalance', 'cardPresent',
       'expirationDateKeyInMatch', 'isFraud']
df=df.reindex(columns=col_names)
```

```python
df.head()
```

```python
#Confirm no duplicated rows

df.duplicated().sum()
```

```python
#There are 5000 unique customer ID's

customers=df.customerId.unique()
len(customers)
```

```python
#14% of last 4 card digits are not of length 4


not_last4=df['cardLast4Digits'].apply(lambda x: len(str(x))!=4).sum()

not_last4_ratio= not_last4/len(df)

print(not_last4_ratio*100)
```
